```python
import pandas as pd

df = pd.read_csv('documents.csv')

print(df)
```

```python
texts = df.text_right
texts = texts.to_numpy()
```

```python
count = dict()
total_number_words = 0
total_words_length = 0
total_unique_words_length = 0

for text in texts:
    words = text.split()
    for word in words:
        if word in count:
            count[word] += 1
        else:
            count[word] = 1
            total_unique_words_length += len(word)
        total_words_length += len(word)
        total_number_words += 1
```

```python
number_docs = 0
total_number_words = 0

for text in texts:
    total_number_words += len(text.split())
    number_docs += 1
    
print("Number of documents: ")
print(number_docs)
print("The average of document length in words is: ")
print(total_number_words/number_docs)
print("Total collection size in words: ")
print(total_number_words)
```

```python
sorted_count = sorted(count.items(), key=lambda x:x[1], reverse = True)
```

```python
sorted_count
```

```python
total_number_unique_words = len(sorted_count)
print("Total number of unique words: ")
print(total_number_unique_words)
print("Average word length: ")
print(total_words_length/total_number_words)
print("Average unique word length: ")
print(total_unique_words_length/total_number_unique_words)
```

```python
file = open('stopwords' , 'r')
Lines = file.readlines()
stopwords = []

for line in Lines:
    stopwords.append(line[:-1])
```

```python
stopwords
```

```python
top30 = sorted_count[:30]
```

```python
top30
```

```python
print("Words in top30 that are also in stopwords list:\n")
is_in = False

for word in top30:
    for stopword in stopwords:
        if (word[0] == stopword):
            print(word[0])
            is_in = True
    if (is_in == False):
        print("This word is not in the stopword list:")
        print(word[0])
        print()
    is_in = False
```

```python
import math
import matplotlib.pyplot as plt

x1 = []
y1 = []
rank = 1

for word in sorted_count:
    x1.append(math.log10(rank))
    y1.append(math.log10(word[1]))
    rank += 1
```

```python
plt.plot(x1, y1)
plt.xlabel('log10 rank')
plt.ylabel('log10 frequency')
plt.title('Zipf\'s law')
plt.show()
```

```python
x2 = []
y2 = []
vocabulary = set()
document = 0

for text in texts:
    document += 1
    words = text.split()
    for word in words:
        if word not in vocabulary:
            vocabulary.add(word)
    x2.append(math.log10(document))
    y2.append(math.log10(len(vocabulary)))
```

```python
plt.plot(x2, y2)
plt.xlabel('log10 documents read')
plt.ylabel('log10 vocabulary')
plt.title('Heaps\' law')
plt.show()
```
